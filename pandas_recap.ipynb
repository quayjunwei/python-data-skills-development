### Combining data

```python
# whenever we cr8 a new df where we want it to be independant from our og df, we use .copy()
# https://nedbatchelder.com/text/names.html
df1 = df.query('Airline == "Southwest Airlines Co."').copy()
df2 = df.query('Airline == "Allegiant Air"').copy()
```

```python
#stack df on top of each other
df_stack = pd.concat([df1,df2])
```

```python
#concatenate based on index values
# default axis is 0
df1 = df1.reset_index(drop = True)
df2 = df2.reset_index(drop = True)
df_side = pd.concat([df1,df2], axis =1)
df_side
```

```python
# see how diff concat methods work
#stack concat stack df on top of each other (same number of columns but combined number rows)
# concat with axis = 1 has merged on the similar index (same number of rows of larger dataframe and double the number of columns)
# concat with axis = 1 is tricky cos now we have multiple columns with the same name
#instead we want to merge df based on similar columns (join)
df1.shape, df2.shape, df_stack.shape, df_side.shape
```

```python
# continuening from the previous example
df1 = df.groupby(['Airline','FlightDate'])['DepDelay'].mean().reset_index()
df2 = df.groupby(['Airline','FlightDate'])['ArrDelay'].mean().reset_index()
```

```python
#take any similar columns between the 2 df and combine on those
df1.merge(df2)
```

```python
#another way to merge
#both merge and pd.merge have optional `on` variable
#if both df have same column names, it will auto put a suffix x and y 
pd.merge(df1,df2, on = ['Airline'])
```

```python
# if column names in both df are different
pd.merge(df1,df2, left_on = ['Airline'], right_on=['Airline'])
```

```python
#change suffix
pd.merge(df1,df2, on = ['Airline'], suffixes=('_dep','_arr'))
```
