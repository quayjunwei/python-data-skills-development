## Geometry operations

### buffering geometries
expand or shrinks a geometry

```python
point = Point(0, 0)
point
```

```python
point_buffer = point.buffer(0.2)
point_buffer
```

```python
point_buffer = point.buffer(1.0)
point_buffer
```

```python
line = LineString([(0, 0), (0, 1), (1, 1)])
line
```

```python
line_buffer = line.buffer(0.2)
line_buffer
```

```python
# 
line_buffer = line.buffer(1.2)
line_buffer
```

```python
# when we buffer a polygon, eg: a sqaure, the larger the buffer, the more curve the corners will be
```

```python
sqaure = Polygon([(0, 0), (1, 0), (1, 1), (0, 1)])
sqaure
```

```python
square.buffer(0.2)
```

```python
square.buffer(1.0)
```

```python
square.buffer(10.0)
```

## set operation on geometries

```python
circle_1 = Point(0, 0).buffer(2)
circle_2 = Point(3, 0).buffer(2)
circle_1
```

```python
circle_1.union(circle_2)
```

```python
circle_1.intersection(circle_2)
```

```python
circle_1.difference(circle_2)
```

```python
circle_2.difference(circle_1)
```
