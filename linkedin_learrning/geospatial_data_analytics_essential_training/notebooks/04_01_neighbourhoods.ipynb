# Explore the administrative boundaries of the NYC neighbourhoods

```python
import geopandas as gpd
```

```python
file_name = '../data/2010 Neighborhood Tabulation Areas (NTAs).geojson'
```

```python
gdf = gpd.read_file(file_name)
print(len(gdf))
gdf.head()
```

```python
gdf.crs
```

```python
gdf.plot()
```

```python
gdf.shape_area.hist()
```

```python
gdf.shape_area[0]
```

```python
print(gdf.shape_area.dtype)  
```

```python
gdf['shape_area'] = gdf['shape_area'].astype(float)
gdf['shape_leng'] = gdf['shape_leng'].astype(float)
```

```python
print(gdf.shape_area.dtype)  
```

```python
gdf.shape_area.hist()
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(1, 2, figsize = (10, 5))

gdf.shape_area.hist(ax=ax[0], bins = 20)
gdf.shape_leng.hist(ax=ax[1], bins = 20)

ax[0].set_yscale('log')
ax[1].set_yscale('log')

ax[0].set_title('NYC neighborhood areas', fontsize=14, pad=16)
ax[1].set_title('NYC neighborhood length', fontsize=14, pad=16)
```
