# Turning tabular data into Geospatial

```python
import pandas as pd
```

```python
tdf = pd.read_csv('../data/2015_Street_Tree_Census_-_Tree_Data_20241223.csv', nrows = 10000)
```

```python
tdf.groupby("health").size()
```

```python
from shapely.geometry import Point
import geopandas as gpd
```

```python
geometry = [Point(xy) for xy in zip(tdf['longitude'], tdf['latitude'])]
geometry[1]
```

```python
gdf = gpd.GeoDataFrame(tdf, geometry=geometry)[['tree_id', 'stump_diam', 'status', 'health', 'geometry']]
```

```python
gdf.head()
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(1,1,figsize=(18,10))
gdf.plot(ax=ax, column ='status', legend = True)
```
